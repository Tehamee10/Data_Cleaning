**Data Cleaning in Pandas**

```python
import pandas as pd
```

```python
df = pd.read_csv('kaggle_survey_2017_2021.csv')
df
```

**Drop Duplicate Rows**

```python
df.drop_duplicates()
```

**Drop Columns With Too Many NaNs**

```python
threshold = len(df) * 0.5
df = df.dropna(thresh=threshold, axis=1)
```

**Drop Completely Empty Rows**

```python
df = df.dropna()
df
```

**Clean Column Names**

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[^\w]', '', regex=True)
```

**Fill Remaining NaNs**

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna('Unknown')
    else:
        df[col] = df[col].fillna(0)
```

**Label Encode Categorical Columns**

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = ['What_is_your_gender_Selected_Choice',
                    'In_which_country_do_you_currently_reside',
                    'What_is_the_highest_level_of_formal_education']

le = LabelEncoder()
for col in categorical_cols:
    if col in df.columns:
        df[col + '_Encoded'] = le.fit_transform(df[col])
```

```python
country_counts = df['q3'].value_counts()
print(country_counts)
```

```python
import matplotlib.pyplot as plt

top5 = country_counts.head(5)

plt.figure(figsize=(8,6))
top5.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.title('Top 5 Countries by Respondent Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Save Cleaned Data**

```python
df.to_csv('cleaned_survey.csv', index=False)
```
